```python
# Creo la clase:
class TiendaOnline:
     # Creo el constructor y declaro los atributos
    def __init__(self):
        self.inventario = []# una lista en la que cada artículo será un diccionario {"nombre",  "precio", "cantidad"}
        self.clientes = {} # un diccionario en el que se guardarán los clientes
        self.historial= {} # diccionario vacio que contiene, {"email y compras"} PTE DE ELIMINAR********************************************************************************
        self.ventas_totales = float(0.0)  # Las ventas, parten de 0.00€

# Método 1: Agregar productos
    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:  # Itero cada diccionario de la lista inventario
            if producto['nombre'] == nombre:   # Si el producto existe en el inventario
                producto['cantidad'] += cantidad # Actualiza su cantidad
                print(f"El producto {nombre} ha sido actualizado a {cantidad} uds.")
                return self.inventario

        else:
            nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad} # Si no está en el inventario
            self.inventario.append(nuevo_producto)
            print(f"Se ha añadido el producto {nombre} al inventario.")

# Método 2: Inventario
    def ver_inventario(self):
        for producto in self.inventario:
            nombre = producto['nombre'] 
            precio = producto['precio']
            cantidad = producto['cantidad']
            print(f"Nombre: {nombre}, Precio: ${precio}, Cantidad: {cantidad}")

# Método 3: Buscar productos
    
    def buscar_producto_por_nombre(self, nombre):
        resultado = [] # Creo una variable en la que guardo la búsqueda para que si no la encuentra, me diga que no hay
        for producto in self.inventario: 
            if producto["nombre"].lower() == nombre.lower(): # Si encuentra el artículo, lo guarda en 'resultado' y lo imprime
                print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                resultado.append(producto)
 
        if not resultado: # Si no se guardó nada en resultado (no apareció el artículo en el inventario)
            print(f"El artículo {nombre}no está en el inventario.")
            
# Método 4: Actualizar Stock
      
    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto['nombre'] == nombre: # Busca por el nombre y suma la cantidad nueva al inventario existente
                producto['cantidad'] += cantidad
                print(f"El inventario de {nombre} se ha actualizado.")  
                return self.inventario
            
        else:
            print(f"El artículo {nombre} no está en el inventario.")

    # Método 5: Eliminar Productos
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre: 
                self.inventario.remove(producto)
                print(f"El artículo {nombre} se ha eliminado.")   # Imprime un mensaje de éxito de la eliminación del producto
                return self.inventario  # Nos devuelve el inventario actualizado.
        else:
            print(f"No se encontró el producto  {nombre}.")

# Método 6: Valor Inventario
    def calcular_valor_inventario(self):
        valor_total = 0   
        for producto in self.inventario: 
            valor_producto = producto['precio'] * producto['cantidad'] 
            valor_total += valor_producto
        print(f"El valor total del inventario es: {valor_total}")
        return f"El valor total del inventario es: {valor_total}"       

# Método 8:Realizar compras  
    def realizar_compra(self):
        carrito = []  # Creo carrito vacío en el que se irán guardando las compras
        seguir_comprando = True  # Por defecto está en sí, pero al final de cada compra le pregunto si quiere seguir con la compra y si dice que no, rompe el while.
        coste_total = 0  # Creo variable en la que se irá sumando el total de cada compra      

        while seguir_comprando == True:  
            print(self.ver_inventario())  # Visualiza el inventario antes de la compra
            nombre_producto = input("¿Qué artículo quieres comprar?").capitalize()
            unidades = int(input("¿Cuántas unidades quieres compar?")) 
    
            for producto in self.inventario:
                if producto['nombre'] != nombre_producto:
                    print('Artículo no encontrado')
                    continue
                elif producto['nombre'] == nombre_producto and producto['cantidad'] <= unidades:
                    print('Cantidad en stock insuficiente')
                    continue
                elif producto['nombre'] == nombre_producto and producto['cantidad'] >= unidades:  # Si existe el artículo y hay unidades suficientes, lo añado al carrito
                    carrito.append({'precio': producto['precio'], 'cantidad': unidades})
                    producto['cantidad'] -= unidades  # Resto las unidades compradas del inventario
                    coste_total += producto['precio'] * unidades  # Calculo el coste total de esta compra
                    print(f'Has añadido al carrtito {unidades} unidades de {nombre_producto}. Total: {coste_total}€')


            seguir = input("Quieres seguir con la compra? (s/n): ")  # Preguntamos al cliente si quiere seguir comprando
            if seguir.lower() != 's':
                seguir_comprando = False  # Si la respuesta no es 's', terminamos la compra

    
        print(f"Coste total de la compra: {coste_total}€")

# Método 9:Procesar Pago
    def procesar_pago(self):
        try:
            total = float(input("Ingrese la cantidad total a pagar: "))
            pago = float(input("Ingrese la cantidad de tu pago: "))
            cambio = pago - total
            if cambio >= 0:
                print(f"Pago con exito. Tu cambio es: ${cambio}")
            else:
                print(f"Error: Cantidad insuficiente, te falta dinero Tacaño.")
        except ValueError:
            print("Error: Ingrese un número  válido para la cantidad y el pago.")

# Método 10:Agregar Clientes
    def agregar_cliente(self, nombre, email):
        if email not in self.clientes: # Si no existe, agrega al cliente al diccionario de clientes con su nombre y correo
            self.clientes[email] = [nombre, email]
            print(f"Cliente {nombre} registrado exitosamente.")
            return self.clientes  # Devuelve el diccionario de clientes actualizado
        else:
            print(f"Error: Este correo electrónico ya está registrado para otro cliente .") 
 
# Método 11: Ver Clientes   
    def ver_clientes(self):
        print("Lista de clientes registrados:")
        for email, clientes in self.clientes.items():
            nombre, email = clientes
            print(f"Nombre: {nombre}, Correo electrónico: {email}") 
    
# Método 12: Registrar Compra 
    def registrar_compra(self, nombre_cliente, carrito):
        if nombre_cliente in self.clientes:  # Comprueba si el cliente está registrado
            total_compra = sum(producto['precio'] * producto['cantidad'] for producto in carrito)  # Calcula el total de la compra con una list comprehension usando for 
            self.ventas_totales += total_compra  # Actualiza las ventas totales de la tienda
            compra_cliente = {"productos": carrito, "total": total_compra}  # Crear un diccionario para la compra del cliente
            
            if nombre_cliente in self.historial:  # Verifica si el cliente ya tiene un historial de compras
                self.historial[nombre_cliente].append({"productos": carrito, "total": total_compra})  # Agrega la compra al historial del cliente
            else:
                self.historial[nombre_cliente] = [compra_cliente]   # Crea un nuevo historial para el cliente
            print(f"Compra registrada para el cliente {nombre_cliente}. Total de compra: €{total_compra}")
        else:
            print(f"Cliente no registrado no se puede realizar la compra")

# Método 13: Ver compras Cliente 
    def ver_compras_cliente(self, email_cliente):
        if email_cliente in self.historial: #Verificar si el cliente esta en el historial
            print(f"Historial de compras para el cliente con correo electrónico {email_cliente}:")
            for compra in self.historial[email_cliente]: #Iteracion sobre cada compra registrada en el historial por cliente
                print("Productos comprados:")
                for producto in compra['productos']: #Iteracion sobre la lista de productos de cada compra
                    print(f" - {producto['nombre']}: {producto['cantidad']} unidades") #Imprimimos el nombre y la cantidad 
                print(f"Total: ${compra['total']}") #Imprimimos la compra total
        else:
            print("El cliente no tiene historial de compras.")  # El cliente no tiene historial       
 # Método 14: Ventas Totales
    def calcular_ventas_totales(self):
        total_ventas = 0  # Inicializa una variable para el total de ventas en 0
        for compras_cliente in self.historial.values(): # Itera sobre los valores del historial de los clientes 
            for compras in compras_cliente:  # Itera sobre cada compra de los clientes
                total_ventas += compras["total"] #Suma el total de las ventas
        print(f"Las ventas totales de la tienda son: ${total_ventas:}")        
        return total_ventas
```
